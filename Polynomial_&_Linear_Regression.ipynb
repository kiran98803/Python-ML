#Visualizing Dummy Data

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

x=np.array([1,2,3,4,5]).reshape(-1,1)
y=np.array([3,6,13,28,55])
```

##Linear Rregression

```python
linear_model=LinearRegression()
linear_model.fit(x,y)
y_pred_linear=linear_model.predict(x)
```

##POLYNOMILA REGRESSION


```python
poly=PolynomialFeatures(degree=2)
x_poly=poly.fit_transform(x)
poly_model=LinearRegression()
poly_model.fit(x_poly,y)
y_pred_poly=poly_model.predict(x_poly)

plt.scatter(x,y,color='red')
plt.plot(x,y_pred_linear,label='linear_fit',color='blue')
plt.plot(x,y_pred_poly,label='Polynomial_fit',color='green')
plt.title('Dummy Data')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

##Real Dataset: Salary Prediction

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

##Load Dataset

```python
df=pd.read_csv("https://raw.githubusercontent.com/Naresh312143/mahalingpur-python-ml/refs/heads/main/Salary_Data.csv")
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

##Train-test split

```python
x=df[["Years of Experience"]]
y=df["Salary"]

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

##Linear Regression

##Train linear regression

```python
linear_model=LinearRegression()
linear_model.fit(x_train,y_train)
y_pred_linear=linear_model.predict(x_test)
```

##Metrices

```python
mse_lin=mean_squared_error(y_test,y_pred_linear)
rmse_lin=np.sqrt(mse_lin)
r2_lin=r2_score(y_test,y_pred_linear)

print("Linear Regression")
print("MSE:",mse_lin)
print("RMSE:",rmse_lin)
print("R2:",r2_lin)
```

#Polynpmial Regression (Degree 2 & 4)

##Degree2

```python
poly2=PolynomialFeatures(degree=2)
x_train_poly2=poly2.fit_transform(x_train)
x_test_poly2=poly2.transform(x_test)

poly2_model=LinearRegression()
poly2_model.fit(x_train_poly2,y_train)
y_pred_poly2=poly2_model.predict(x_test_poly2)
```

```python
x_train_poly2
```

```python
x_test_poly2
```

##Metrices

```python
mse_poly2=mean_squared_error(y_test,y_pred_poly2)
rmse_poly2=np.sqrt(mse_poly2)
r2_poly2=r2_score(y_test,y_pred_poly2)
```

##Degree 4

```python
poly4=PolynomialFeatures(degree=4)
x_train_poly4=poly4.fit_transform(x_train)
x_test_poly4=poly4.transform(x_test)

poly4_model=LinearRegression()
poly4_model.fit(x_train_poly4,y_train)
y_pred_poly4=poly4_model.predict(x_test_poly4)
```

```python
x_train_poly4
```

```python
x_test_poly4
```

##Metrices

```python
mse_poly4=mean_squared_error(y_test,y_pred_poly4)
rmse_poly4=np.sqrt(mse_poly4)
r2_poly4=r2_score(y_test,y_pred_poly4)

print("\nPolynomial Regression (Degree 2)")
print("MSE:",mse_poly2)
print("RMSE:",rmse_poly2)
print("R2:",r2_poly2)

print("\nPolynomial Regression (Degree 4)")
print("MSE:",mse_poly4)
print("RMSE:",rmse_poly4)
print("R2:",r2_poly4)
```

#Visualization of All Models

Generate smooth line for predictions

```python
x_range=np.linspace(x.min(),x.max(),100).reshape(-1,1)

#Create Dataframe With column name
x_range_df=pd.DataFrame(x_range,columns=['Years of Experience'])

#Transform using polynomial features
x_range_poly2=poly2.transform(x_range_df)
x_range_poly4=poly4.transform(x_range_df)

#Predict using models
y_range_linear=linear_model.predict(x_range_df)
y_range_poly2=poly2_model.predict(x_range_poly2)
y_range_poly4=poly4_model.predict(x_range_poly4)

#Plot
plt.scatter(x,y,color='red',label='Actual Data')
plt.plot(x_range,y_range_linear,color='blue',label='Linear_fit')
plt.plot(x_range,y_range_poly2,color='green',label='Polynomial_fit_degree2')
plt.plot(x_range,y_range_poly4,color='orange',label='Polynomial_fit_degree4')
plt.title('Salary Prediction')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.grid(True)
```

#Model performance Comparison

```python
metrics_df=pd.DataFrame({
    'Model':['Linear','Polynomial (Degree 2)','Polynomial (Degree 4)'],
    'MSE':[mse_lin,mse_poly2,mse_poly4],
    'RMSE':[rmse_lin,rmse_poly2,rmse_poly4],
    'R2':[r2_lin,r2_poly2,r2_poly4]
})

metrics_df.style.format({
    'MSE':'{:.2f}',
    'RMSE':'{:.2f}',
    'R2':'{:.4f}'
})
```
